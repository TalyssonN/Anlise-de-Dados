<h1>Análise sobre o Desempenho de Alunos em Exames</h1>

<h2>Questões a serem respondidas:

<h3>1.Relação entre hábitos de estudo e desempenho:
   
Alunos que têm um padrão de sono regular (quantidade de horas dormidas por noite) têm um desempenho superior em relação aos que dormem pouco?

Alunos que participam de atividades extracurriculares apresentam uma performance acadêmica superior ou inferior aos que não participam?

2.Influência do envolvimento dos pais:

Existe uma relação positiva entre o envolvimento dos pais (participação em reuniões escolares, acompanhamento de lição de casa) e o desempenho acadêmico dos alunos?

</h3>

</h2>

```python
import pandas as pd;
import numpy as np;
import matplotlib.pyplot as plt;
import seaborn as sns;
plt.style.use('ggplot')
```

```python
df = pd.read_csv('StudentPerformanceFactors.csv')
df.head(10)
```

<h3>Adquirindo informações das colunas.</h3>

```python
df.info()
```

<h3>
    1.Relação entre hábitos de estudo e desempenho:
    
    Existe uma correlação entre o número de horas dormidas e a pontuação nos exames?
</h3>

<h4>Selecionar as horas dormidas e os resultados dos exames.</h4>

```python
HrsDormidas = df.loc[df['Sleep_Hours'], ['Sleep_Hours', 'Previous_Scores', 'Exam_Score']]
print(HrsDormidas)
```

```python
plt.figure(figsize = (12, 5));
sns.lineplot(data=HrsDormidas, x='Sleep_Hours', y='Exam_Score')
plt.title('Média das Pontuações Anteriores por Faixa de Horas de Sono')
plt.xlabel("Horas de Sono")
plt.ylabel("Pontuação")
plt.xticks(rotation=30)
plt.show()
```

<h4>De acordo com o gráfico mostrado acima, os alunos que possuem mais horas de sono conseguem alcançar uma pontuação maior nos exames, visto que estão descansados e com a mente mais calma!</h4>

<h3>
    1.Relação entre hábitos de estudo e desempenho:
    
    Alunos que participam de atividades extracurriculares apresentam uma performance acadêmica superior ou inferior aos que não participam?
</h3>

```python
df = pd.read_csv('StudentPerformanceFactors.csv')
df.head(10)
```

<h4>
    Filtrar os alunos que possuem atividades extracurriculares e seus resultados nos exames, antes e depois de iniciarem as atividades.</h4>

```python
select_atvEx = df[df['Extracurricular_Activities'] == 'Yes'][['Previous_Scores','Extracurricular_Activities', 'Exam_Score']]
display(select_atvEx)
```

<h4>Média das notas finais dos alunos que realizam atividades extras.</h4>

```python
mediaNotas = select_atvEx['Exam_Score'];
print(mediaNotas.mean());
```

<h4>Selecionar os alunos que estão com as notas abaixo da média e que realizam atividades extracurriculares.</h4>

```python
notaBaixa = select_atvEx[select_atvEx['Exam_Score'] < 67.5]
display(notaBaixa)
```

<h4>Selecionar os alunos que não realizam atividades extras.</h4>

```python
select_semAtvEx = df[df['Extracurricular_Activities'] == 'No'][['Previous_Scores','Extracurricular_Activities', 'Exam_Score']]
display(select_semAtvEx)
```

```python
mediasemAtvEx = select_semAtvEx['Exam_Score'];
print(mediasemAtvEx.mean());
```

```python
notasemAtvEx = select_semAtvEx[select_semAtvEx['Exam_Score'] < 66.9]
print(notasemAtvEx)
```

<h5>
    
    Alunos com atividade extra: 3938
    
    Alunos com atividade extra e com notas baixas: 2081
    Total: 1.857
    
    Alunos sem atividade extra: 2669
    Alunos sem atividade extra e com notas baixas: 1244
    Total: 1.425
</h5>

```python

# Definir os dados
categories = ['Com atividade extra', 'Sem atividade extra']
students_with_activity = [3938, 2669]
students_with_activity_low_grades = [2081, 1244]

# Posições no eixo x
x = np.arange(len(categories))

# Largura das barras
width = 0.2

# Criar as barras
fig, ax = plt.subplots(figsize=(10, 6))

bar1 = ax.bar(x - width, students_with_activity, width, label='Total de Alunos')
bar2 = ax.bar(x, students_with_activity_low_grades, width, label='Com Notas Baixas')

# Adicionar rótulos e título
ax.set_xlabel('Atividade Extracurricular')
ax.set_ylabel('Número de Alunos')
ax.set_title('Distribuição de Alunos com e sem Atividades Extracurriculares e Notas Baixas')

# Definir os ticks no eixo x
ax.set_xticks(x)
ax.set_xticklabels(categories)

# Adicionar legenda
ax.legend()

# Mostrar gráfico
plt.tight_layout()
plt.show()
```

<h3>
    
    De acordo a análise dos dados, temos uma diferença muito grande na quantidade de alunos que possuem atividades extras, contudo, o maior indice comparativo de notas baixas está nos alunos sem atividades extras, visto que a quantidade de alunos que realizam atividades extras e tem boas notas, está acima do número de alunos que não tem nenhuma atividade além da escola e estão com notas baixas. É necessário analisar qual as atividades que estão sendo realizadas por os alunos, fora do ambiente escolar, para determinar as mesmas, atrapalham ou estimulam os alunos a estudarem.
</h3>

<h3>
    2.Influência do envolvimento dos pais:
    
    
Existe uma relação positiva entre o envolvimento dos pais (participação em reuniões escolares, acompanhamento de lição de casa) e o desempenho acadêmico dos alunos?
</h3>

```python
df = pd.read_csv('StudentPerformanceFactors.csv')
df.head(10)
```

```python
media = df['Exam_Score'].mean()
print(media)
```

```python
acima_da_media = df[(df['Parental_Involvement'] == 'High') & (df['Exam_Score'] > 67.2)][['Previous_Scores','Parental_Involvement','Exam_Score']]
display(acima_da_media)
```

```python
acima_da_media = df[(df['Parental_Involvement'] == 'Medium') & (df['Exam_Score'] > 67.2)][['Previous_Scores','Parental_Involvement','Exam_Score']]
display(acima_da_media)
```

```python
acima_da_media = df[(df['Parental_Involvement'] == 'Low') & (df['Exam_Score'] > 67.2)][['Previous_Scores','Parental_Involvement','Exam_Score']]
display(acima_da_media)
```

```python
plt.figure(figsize = (12, 5));
sns.lineplot(data=df, x='Parental_Involvement', y='Exam_Score')
plt.title('Média das Pontuações dos Alunos por Grau de Envolvimento dos Pais')
plt.xlabel("Horas de Sono")
plt.ylabel("Pontuação")
plt.xticks(rotation=30)
plt.show()
```

<h4>
    
    O número de alunos que possuem o índice de envolvimento parental no alto e médio tem uma diferença muito grande em relação aos que tem pouco envolvimento dos pais na educação.
    
    Dados:
    Alto: 1046;
    Médio: 1490;
    Baixo: 472;
    
    Diante de tais valores e do gráfico apresentado, fica evidente a forte influência que os pais exercem na educação dos filhos.
</h4>
